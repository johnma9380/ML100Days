```python
import pandas as pd
import numpy as np

from scipy import stats

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.manifold import TSNE
```

## 讀取檔案

```python
""" Your code hear: """
train = pd.read_csv('./train.csv')
train
```

## 取出連續型特徵 (方便降維)

```python
""" Your code hear: """
c_cols = [n for t, n in zip(train.dtypes, train.columns) if t != 'object' and n not in ['Id', 'SalePrice']]
    
c_cols
```

## 分析缺失值

```python
""" Your code hear: 練習印出缺失值圖表 """
missing = train.isnull().sum(axis = 0)
missing = missing[missing > 0]
missing.sort_values(inplace=True)

plt.figure(figsize = (8, 4))
plt.bar(list(missing.index), np.array(missing))
plt.xticks(rotation='vertical')
plt.show()
print(missing)
```

## 去除缺失值

```python
""" Your code hear: 去除缺失值 """
train[['GarageYrBlt','LotFrontage','MasVnrArea']].isnull()
df = train[train['GarageYrBlt'].notnull()]
df = df[df['LotFrontage'].notnull()]
df = df[df['MasVnrArea'].notnull()]
df
```

## 轉化成 ndarray 後進行標準化，並將 Y 透過四分位數轉換成四個類別

```python
""" Your code hear: """
min_max_salePrice = (df['SalePrice'] - df['SalePrice'].min() ) / (df['SalePrice'].max() - df['SalePrice'].min() )
df['min_max_salePrice'] = min_max_salePrice
```

```python
arr = np.array(min_max_salePrice, dtype = 'float')
bins = stats.mstats.mquantiles(arr).tolist()
bins = [0] + bins + [df['SalePrice'].max() + 1]
print(bins)
```

```python
score_cut = pd.cut(arr, bins, labels=False, retbins=True, right=False)
salePrice_cateor = score_cut[0].tolist()
df['SalePrice_Categories'] = salePrice_cateor
df[['SalePrice', 'SalePrice_Categories','min_max_salePrice']]
```

```python
features = c_cols + ['SalePrice']
corr, p_val = stats.spearmanr(df[features], nan_policy = 'omit')

# 視覺化相關性 和 p-value
plt.figure(figsize = (20, 8), dpi = 100)
plt.subplot(1, 2, 1)
plt.title('Spearman correlation heatmap')
sns.heatmap(corr, xticklabels = features, yticklabels = features) # 相關性

plt.subplot(1, 2, 2)
plt.title('P Value heatmap')
sns.heatmap(p_val, xticklabels = features, yticklabels = features) # P-Value
plt.show()
```

## 利用 TSNE 降維

```python
plt.figure()
# for i, feature in enumerate(['tsne1', 'tsne2']):
plt.plot(df['SalePrice'], 'o')
plt.show()
plt.plot(df['min_max_salePrice'], 'o')
plt.show()

plt.scatter(df['SalePrice'], df['min_max_salePrice'])
plt.show()
```

```python
X = np.array(df[['SalePrice_Categories','SalePrice']], dtype = 'float')
tsne = TSNE(
    n_components=2, init='pca', perplexity = 40, 
    early_exaggeration = 8, learning_rate = 150,
    n_iter = 2000
)
X_tsne = tsne.fit_transform(X)

df['tsne1'] = [i[0] for i in X_tsne]
df['tsne2'] = [i[1] for i in X_tsne]
```

```python
plt.figure()
# for i, feature in enumerate(['tsne1', 'tsne2']):
plt.plot(df['tsne1'], 'o')
plt.plot(df['tsne2'], 'o')
plt.show()

plt.scatter(df['tsne1'], df['tsne2'])
plt.show()
```

```python
X = np.array(df[['SalePrice_Categories','min_max_salePrice']], dtype = 'float')
tsne = TSNE(
    n_components=2, init='pca', perplexity = 40, 
    early_exaggeration = 8, learning_rate = 150,
    n_iter = 2000
)
X_tsne = tsne.fit_transform(X)


df['tsne1'] = [i[0] for i in X_tsne]
df['tsne2'] = [i[1] for i in X_tsne]
```
