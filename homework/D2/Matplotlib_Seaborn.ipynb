```python
import pandas as pd
import numpy as np

from scipy import stats

import matplotlib.pyplot as plt
import seaborn as sns
```

# Matplotlib 教學
## 方法一

```python
# y = sin(x)
x1 = np.linspace(0, 5 * np.pi, 200)
y1 = np.sin(x1)

#  y = cos(x)
x2 = np.linspace(0, 5 * np.pi, 200)
y2 = np.cos(x2)

# 初始化一個 (row, col) = (8, 10), dpi = 100 的 figure
fig = plt.figure(figsize=(5, 5), dpi = 100)

# 在 fig 上初始化兩個子圖，排列方式為 (row, col) = (2, 1)
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

# 開始做圖
ax1.plot(x1, y1)
ax2.plot(x2, y2)

# plt.plot(), plt.hist(), ...等函數會在最近的 ax 做圖
# plt.plot(x1, y1)

# 將 buffer 中的圖表輸出至螢幕
plt.show()
```

## 方法二

```python
# 透過 subplots 生成 figure 及內部的 axes (axes.shape = (2, 2))
fig, axes = plt.subplots(2, 2)
# 設定 figure 大小及解析度
fig.set_size_inches(10, 6)
fig.set_dpi(100)

# 繪圖
for i in range(2):
    for j in range(2):
        axes[i, j].plot(x1, y1)
        axes[i, j].title.set_text(f'ax{i * 2 + j + 1}')

# 調整子圖間的距離，下面兩者等價，同學可以告訴我原因嗎？
# plt.subplots_adjust(wspace=0.2, hspace=0.4)
fig.subplots_adjust(wspace=0.2, hspace=0.4)

plt.show()
```

## 方法三

```python
plt.figure(figsize = (10, 6), dpi = 100)

for i in range(4):
    plt.subplot(2, 2, i + 1, title = f'ax{i + 1}')
    plt.plot(x1, y1)

plt.subplots_adjust(wspace=0.2, hspace=0.4)
plt.show()
```

```python
# 儲存 figure
fig.savefig(fname = 'test.png', dpi = 100)
```

# Seaborn 教學

```python
fig = plt.figure(figsize = (10, 5))

points = np.random.randn(200).cumsum()

ax1 = plt.subplot(1, 2, 1, title = 'bar')
sns.histplot(data = points, ax = ax1)

ax2 = plt.subplot(1, 2, 2, title = 'distribution')
sns.histplot(data = points, kde = True, ax = ax2)
# !!!
# sns.displot(data = points, kde = True, ax = ax2)
# !!!

plt.show()
```
