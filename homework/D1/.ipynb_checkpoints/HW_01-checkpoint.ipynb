## 1. 使用 Numpy, Pandas 分別計算 “homework.csv” 中每個 feature 的平均值、中位數、最大值、最小值，並比較兩者運算時間

```python
# -------------- Numpy --------------
arr = np.array(df, dtype = 'float')

start = time.time()
""" Your Code Here: 用 Numpy 計算 “homework.csv” 中每個 feature 的平均值、中位數、最大值、最小值，並比較兩者運算時間"""
print(np.mean(arr, axis = 0))
print(np.median(arr, axis = 0))
print(np.max(arr, axis = 0))
print(np.min(arr, axis = 0))
end = time.time()
print(f'Numpy has done in {(end - start):.4f} sec.')

# -------------- Pandas --------------
start = time.time()
""" Your Code Here: 用 Pandas 計算 “homework.csv” 中每個 feature 的平均值、中位數、最大值、最小值，並比較兩者運算時間"""
print(df.mean())
print(df.median())
print(df.max())
print(df.min())
end = time.time()
print(f'Pandas has done in {(end - start):.4f} sec.')
```

## 2. 利用以下方式篩選離群值(option)：
* feature 的平均值組成一個高維向量 Vec
* 計算每個 sample 與 Vec 的距離
* 計算距離的平均值、標準差，將距離平均值超過三個標準差的點當作離群值刪除
* __請將刪除離群值後的 dataframe 取名為 df_after_delete__

```python
# 計算各個 sample 與 Vec 之距離
""" Your Code Here: 計算各個 sample 與 Vec 之距離"""
for i,n in df.mean().items():
    df[i + '_avg'] = n

for i in df.columns:
    if '_avg' not in i:
        df[i+'_minus'] = df[i] - df[i + '_avg']

df_minus = df.filter(regex='_minus$', axis=1)


# 計算 distance 的 mean, std, 並篩選離群值
""" Your Code Here: 計算 distance 的 mean, std, 並篩選離群值"""
for i,n in df.filter(regex='_minus$', axis=1).mean().items():
    df[i + '_avg'] = n
for i,n in df.filter(regex='_minus$', axis=1).std().items():
    df[i + '_std'] = n

for i in df.filter(regex='_minus$', axis=1).columns:
    df[i + '_score'] = abs(df[i] - df[i + '_avg'] / df[i + '_std'])

decdf = df.filter(regex='_score$', axis=1) < 3

isNorValue = []
isNotNorValue = []

for i in range(0, len(decdf)):
    isNorValue.append(all(decdf[i:i+1].values[0]))
df[isNorValue]


# 刪除離群值
""" Your Code Here: 刪除離群值"""
isNotNorValue = []
for index, item in enumerate(isNorValue):
    if item == False:
        isNotNorValue.append(index)
df_after_delete = df.drop(isNotNorValue)   
```
